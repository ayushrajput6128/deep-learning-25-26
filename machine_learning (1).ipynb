```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("/content/Building_Permits.csv")
```

```python
df.replace("?", np.nan, inplace=True)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
df= pd.read_csv("/content/tvmarketing - tvmarketing.csv")
sales = np.array([54,44,45,65,875,76,]).reshape(-1,1)
TV = np.array([184,166,79,97,965,735]).reshape(-1,1)
model = LinearRegression()
model.fit(TV,sales)
m=model.coef_[0]
b=model.intercept_
print(m,b)
model_pred=model.predict(sales)
plt.scatter(TV,sales,color="Green")
plt.plot(TV,model_pred,color='red',label="Linear Regression")
plt.xlabel("TV")
plt.ylabel("sales")
plt.legend()
plt.show()
```
